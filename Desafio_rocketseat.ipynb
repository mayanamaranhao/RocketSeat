**Bibliotecas**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```



**Tratamento e formatação dos dados**



```python
# Dados fornecidos
dados_vendas = {
    'mes': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'vendas': [2000, 2200, 2300, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300]
}
```

```python
# Criando o DataFrame
df_vendas = pd.DataFrame(dados_vendas)
```

```python
# Verificando os dados
print(df_vendas)
```

**Treinamento do modelo de regressão linear**


```python

# Transformando os meses em números para treinar o modelo
df_vendas['mes_numero'] = np.arange(1, 13)  # Janeiro = 1, Fevereiro = 2, ..., Dezembro = 12

# Separando os dados
X = df_vendas[['mes_numero']]  # Variável independente
y = df_vendas['vendas']        # Variável dependente (vendas)

# Treinando o modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Prevendo as vendas para dezembro (mês 12), agora com o nome da coluna correto
previsao_dezembro = modelo.predict(pd.DataFrame({'mes_numero': [12]}))
print(f'Previsão de vendas para Dezembro: {previsao_dezembro[0]:.2f}')
```

**Montar um histograma dos dados de vendas**

```python
# Plotando o histograma das vendas
plt.figure(figsize=(8, 5))
sns.histplot(df_vendas['vendas'], bins=10, kde=True)
plt.title('Histograma das Vendas')
plt.xlabel('Vendas')
plt.ylabel('Frequência')
plt.show()
```

**Montar um scatter plot**

```python
# Plotando o scatter plot das vendas por mês
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_vendas['mes_numero'], y=df_vendas['vendas'])
plt.title('Scatter Plot das Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.show()
```
