```python
# Importar pandas
import pandas as pd
```

```python
# Carregar os dados
df = pd.read_csv('/content/netflix daily top 10.csv')
```

```python
# Tipos de dados disponíveis
print("Tipos de dados disponíveis:")
print(df.dtypes)
```

```python

# Período da análise feita
periodo = df['As of'].unique()
print("\nPeríodo da análise feita:\n", periodo)
```

```python
# Tamanho da base de dados
tamanho_base = df.shape[0]
print("Tamanho da base de dados:", tamanho_base)
```

```python
# Verificar dados nulos
dados_nulos = df.isnull().sum()
print("\nDados nulos por coluna:")
print(dados_nulos)

```

```python
# Analisar outliers na coluna "Viewership Score"
# Para simplificar, consideraremos um outlier qualquer valor acima de 90 como potencial
outliers = df[df['Viewership Score'] > 90]
print("\nPossíveis outliers na coluna 'Viewership Score':")
print(outliers[['Title', 'Viewership Score']])
```
